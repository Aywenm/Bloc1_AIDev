```python
# importing the libraries
import requests
import time
import json
import os

# setting the api key
api_key = os.getenv("VT_API_KEY")
headers = {'x-apikey': api_key}
```

```python
# List of domains to query
domains = [
    "google.com",
    "youtube.com",
    "facebook.com",
    "amazon.com",
    "wikipedia.org",
    "twitter.com",
    "instagram.com",
    "linkedin.com",
    "netflix.com",
    "bing.com",
    "yahoo.com",
    "ebay.com",
    "reddit.com",
    "pinterest.com",
    "office.com",
    "live.com",
    "msn.com",
    "craigslist.org",
    "wordpress.com",
    "tumblr.com",
    "imdb.com",
    "espn.com",
    "bbc.com",
    "cnn.com",
    "nytimes.com",
    "forbes.com",
    "huffpost.com",
    "quora.com",
    "github.com",
    "stackoverflow.com",
    "dropbox.com",
    "salesforce.com",
    "adobe.com",
    "medium.com",
    "vimeo.com",
    "soundcloud.com",
    "aliexpress.com",
    "booking.com",
    "tripadvisor.com",
    "indeed.com",
    "glassdoor.com",
    "yelp.com",
    "flickr.com",
    "slack.com",
    "zoom.us",
    "skype.com",
    "whatsapp.com",
    "paypal.com",
    "stripe.com",
    "uber.com",
    "lyft.com",
    "airbnb.com",
    "kickstarter.com",
    "etsy.com",
    "roblox.com",
    "twitch.tv",
    "disneyplus.com",
    "hulu.com",
    "apple.com",
    "microsoft.com",
    "oracle.com",
    "sap.com",
    "ibm.com",
    "intel.com",
    "amd.com",
    "nvidia.com",
    "tesla.com",
    "walmart.com",
    "target.com",
    "costco.com",
    "ikea.com",
    "homedepot.com",
    "lowes.com",
    "bestbuy.com",
    "cnet.com",
    "techcrunch.com",
    "wired.com",
    "theverge.com",
    "mashable.com",
    "gizmodo.com",
    "engadget.com",
    "arstechnica.com",
    "reuters.com",
    "bloomberg.com",
    "economist.com",
    "financialtimes.com",
    "wsj.com",
    "dailymail.co.uk",
    "mirror.co.uk",
    "metro.co.uk",
    "foxnews.com",
    "msnbc.com",
    "nbcnews.com",
    "abcnews.go.com",
    "cbsnews.com",
    "usatoday.com",
    "latimes.com",
    "chicagotribune.com",
    "sfgate.com",
    "bbc.co.uk"
]
# Initialize an empty list to store the results
results = []

# Loop through each domain and query VirusTotal
for domain in domains: 
    url = f"https://www.virustotal.com/api/v3/domains/{domain}" # URL to query VirusTotal for domain information
    response = requests.get(url, headers=headers) # Send GET request to the API
    if response.status_code == 200: # Check if the request was successful	
        data = response.json() # Parse the JSON response
        results.append(data) # Append the response to the results list
        print(f"Data for {domain} collected.") # Print a message to indicate success
    else: # If the request was not successful
        print(f"Error for {domain}: {response.status_code} - {response.text}") # Print an error message
    # Sleep to respect the API rate limit (adjust the delay as needed)
    time.sleep(15) # Sleep for 15 seconds

# Save aggregated results to a JSON file for later processing
with open("vt_domains_results.json", "w", encoding="utf-8") as outfile: # Open a file to write the results
    json.dump(results, outfile, indent=2) # Write the results to the file

print("Data collection complete. Aggregated results saved to vt_domains_results.json") # Print a message to indicate completion
```
