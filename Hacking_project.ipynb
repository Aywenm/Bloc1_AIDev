```python
#Importing the required libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd

#URL of the page to be scraped
url = 'https://nvd.nist.gov/vuln/categories' #URL of the page to be scraped
response = requests.get(url) #Requesting the page
soup = BeautifulSoup(response.text, 'html.parser') #Parsing the page
#Finding the table in the page
table = soup.find('table') #Finding the table in the page
#Extracting the data from the table
cwe_id = [] #List to store the CWE ID
cwe_name = [] #List to store the CWE Name
cwe_description = [] #List to store the CWE Description
counter = 0 #Counter to keep track of the columns
#Looping through the table to extract the data
for td in table.find_all('td'):
    #Printing the data
    counter += 1
    print("-----------------") 
    print(td.text) 
    print(counter)
    #Storing the data in the respective lists 
    if counter == 1: 
        cwe_id.append(td.text)
    elif counter == 2: 
        cwe_name.append(td.text)
    elif counter == 3:
        cwe_description.append(td.text)
        counter = 0
    if counter == 3:
            counter = 0
        
```

```python
# Creating a DataFrame from the extracted data
df = pd.DataFrame({'CWE ID': cwe_id, 'CWE Name': cwe_name, 'CWE Description': cwe_description}) #Creating a DataFrame from the extracted data
df.to_csv('cwe.csv', index=False, encoding='utf-8') #Saving the DataFrame to a CSV file

df.head()
df.shape
```
