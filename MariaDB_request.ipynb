```python
import json
import mariadb
import sys
import os
```

```python
# Connect to your MariaDB database
try:
    conn = mariadb.connect(
        user= os.getenv("USER_MARIADB"),
        password=os.getenv("PASSWORD_MARIADB"),
        host=os.getenv("HOST_MARIADB"),
        port=int(os.getenv("PORT_MARIADB")),
        database=os.getenv("DATABASE_MARIADB")
    )
except mariadb.Error as e:
    print(f"Error connecting to MariaDB: {e}")
    sys.exit(1)

cursor = conn.cursor()

# Create a request to get all the data from the table domains
cursor.execute("SELECT * FROM domains")

# Fetch all the data
rows = cursor.fetchall()

# Create a list to store the data
data = []

# Loop through the data and append it to the list
for row in rows:
    data.append({
        "id": row[0],
        "domain": row[1],
        "status": row[2],
        "threat": row[3]
    })
    
# Print the data
print(json.dumps(data, indent=2))
```
